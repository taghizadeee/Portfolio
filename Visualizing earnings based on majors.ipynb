In this project, we'll try to figure out the possible correlations and relationships between the earnings and the university majors. 

```python
import pandas as pd
df=pd.read_csv("https://raw.githubusercontent.com/fivethirtyeight/data/master/college-majors/recent-grads.csv")
print(df.head(8))
```

To drop the missing values from the dataset:


```python
df= df.dropna()
```

```python
import matplotlib.pyplot as plt
% matplotlib inline
df.plot(x='Sample_size', y='Median', kind='scatter')
plt.show()
```

The plot above depicts the relationship between the sample size of a full-time and median salary. Next, we can also visualize the same relationship between the sample size and unemployment rate:

```python
df.plot(x='Sample_size', y='Unemployment_rate', kind='scatter')
```

... and between full-time employment size and median salary and share of women and the unemployment rate:

```python
df.plot(x='Full_time', y='Median', kind='scatter')
df.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter')
```

and same as latter plot, but with men:

```python
df.plot(x='Men', y='Median', kind='scatter')
```

Correlation between size of women and median salary:

```python
df.plot(x='Women', y='Median', kind='scatter')
```

Now let's visualize with histograms to explore the distribution of values in a column. 

```python
cols = ["Sample_size", "Median", "Employed", "Full_time", "ShareWomen", "Unemployment_rate", "Men", "Women"]

fig = plt.figure(figsize=(5,12))
for r in range(1,5):
    ax = fig.add_subplot(4,1,r)
    ax = df[cols[r]].plot(kind='hist', rot=40)
```

One of the efficient ways to visualize the realtionships is the combination of histogram and scatter plot, which is called Scatter Matrix Plot - it enables the exploration of potential relationships and distributions simultaneously. 

```python
from pandas.tools.plotting import scatter_matrix
scatter_matrix(df[['Sample_size', 'Median']], figsize=(6,6))
```

```python
scatter_matrix(df[['Sample_size', 'Median', 'Unemployment_rate']], figsize=(10,10))
```
